```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download('stopwords')
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df = df.fillna('')
```

```python
df['content'] = df['author']+' '+df['title']
```

```python
X =df.drop(columns='label',axis=1)
Y = df['label']
```

stemming reducing a word to its root word

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
df['content'] =df['content'].apply(stemming)
```

```python
X =df['content'].values
Y =df['label'].values
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
model =LogisticRegression()
```

```python
model.fit(X,Y)
```

```python
X_predict =model.predict(X)
training_accuracy =accuracy_score(X_predict,Y)
```

```python
print(training_accuracy)
```

```python
input_data = pd.read_csv('/content/test.csv')
output_data = pd.read_csv('/content/submit.csv')
input_data = input_data.fillna('')
input_data['content'] = input_data['author']+' '+input_data['title']
input_data['content'] =input_data['content'].apply(stemming)
X_test =input_data['content'].values
Y_test =output_data['label'].values
vectorizer.fit(X_test)
X_test = vectorizer.transform(X_test)
model.fit(X_test,Y_test)
X_predict =model.predict(X_test)
testing_accuracy =accuracy_score(X_predict,Y_test)
print(testing_accuracy)

```
